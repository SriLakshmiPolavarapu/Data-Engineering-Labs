```python
rows = soup.find_all('tr')
print(rows[:10])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
print(df1)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
col_labels = soup.find_all('th')
```

```python
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head(10)
print(df3)
```

```python
df1.head(10)
print(df1)
```

```python
frames = [df3, df1]
```

```python
df4 = pd.concat(frames)
df4.head()
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df5.info()
```

```python
df5.shape
```

```python
df6 = df5.dropna(axis=0, how='any')
df6.head()
```

```python
df7 = df6.drop(df6.index[0])
df7.head()
```

```python
df7.rename(columns={'[Place': 'Place'},inplace=True)
df7.head()
```

```python
df7.rename(columns={' Team]': 'Team'},inplace=True)
df7.head()
```

```python
df7['Team'] = df7['Team'].str.strip(']')
df7.head()
```

```python
time_list = df7[' Gun Time'].tolist()
```

```python
time_mins = []

for i in time_list:
    try:
        parts = i.strip().split(':')
        if len(parts) == 2:
            m, s = parts
        elif len(parts) == 3:
            _, m, s = parts  # ignore the hour part if it's "00:mm:ss"
        else:
            raise ValueError(f"Bad format: {i}")

        total = (int(m) * 60 + int(s)) / 60
        time_mins.append(round(total, 2))

    except Exception as e:
        print(f"Skipping '{i}' due to error: {e}")
        time_mins.append(None)
```

```python
df7['Runner_mins'] = time_mins
df7.head()
```

```python
df7.describe(include=[np.number])
```
